```python
import csv

# File name to store recipes
file_name = 'recipes.csv'

# Ensure the CSV file exists
def initialize_csv():
    try:
        with open(file_name, 'x', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["RecipeID", "Name", "Ingredients", "Instructions"])
    except FileExistsError:
        pass

# Function to add a recipe
def add_recipe():
    recipe_id = input("Enter Id: ")
    name = input("Enter Recipe Name: ")
    ingredients = input("Enter Ingredients (comma-separated): ")
    instructions = input("Enter Instructions: ")

    with open(file_name, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([recipe_id, name, ingredients, instructions])
    print("Recipe added successfully!\n")

# Function to view all recipes
def view_recipes():
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        recipes = list(reader)

        if recipes:
            for recipe in recipes:
                print(f"Recipe ID: {recipe[0]}\nName: {recipe[1]}\nIngredients: {recipe[2]}\nInstructions: {recipe[3]}\n---")
        else:
            print("No recipes found.\n")

# Function to search for a recipe by name
def search_recipe():
    name = input("Enter Recipe Name to search: ").lower()

    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        found = False
        for recipe in reader:
            if name in recipe[1].lower():
                print(f"Recipe ID: {recipe[0]}\nName: {recipe[1]}\nIngredients: {recipe[2]}\nInstructions: {recipe[3]}\n---")
                found = True

        if not found:
            print("Recipe not found.\n")

# Function to delete a recipe by ID
def delete_recipe():
    recipe_id = input("Enter Recipe ID to delete: ")
    recipes = []
    deleted = False

    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)
        recipes = list(reader)

    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        for recipe in recipes:
            if recipe[0] != recipe_id:
                writer.writerow(recipe)
            else:
                deleted = True

    if deleted:
        print("Recipe deleted successfully!\n")
    else:
        print("Recipe ID not found.\n")

# Function to update a recipe by ID
def update_recipe():
    recipe_id = input("Enter Recipe ID to update: ")
    recipes = []
    updated = False

    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)
        recipes = list(reader)

    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        for recipe in recipes:
            if recipe[0] == recipe_id:
                print("Enter new details (leave blank to keep current):")
                name = input(f"Name [{recipe[1]}]: ") or recipe[1]
                ingredients = input(f"Ingredients [{recipe[2]}]: ") or recipe[2]
                instructions = input(f"Instructions [{recipe[3]}]: ") or recipe[3]
                writer.writerow([recipe_id, name, ingredients, instructions])
                updated = True
            else:
                writer.writerow(recipe)

    if updated:
        print("Recipe updated successfully!\n")
    else:
        print("Recipe ID not found.\n")

# Main menu
def main():
    initialize_csv()
    while True:
        print("\nRecipe Management System")
        print("1. Add Recipe")
        print("2. View Recipes")
        print("3. Search Recipe")
        print("4. Delete Recipe")
        print("5. Update Recipe")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_recipe()
        elif choice == '2':
            view_recipes()
        elif choice == '3':
            search_recipe()
        elif choice == '4':
            delete_recipe()
        elif choice == '5':
            update_recipe()
        elif choice == '6':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
if __name__ == "__main__":
    main()
```
